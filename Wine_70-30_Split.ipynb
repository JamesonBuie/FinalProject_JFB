```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=pd.read_csv('wine.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
# Changing dependent variable to binary
df['quality'] = df['quality'].replace({'good': 1, 'bad': 0})
```

```python
df.head(10)
```

# Splitting the Data Into Test and Training Set

```python
x2=df[['fixed acidity','volatile acidity','citric acid','residual sugar',
      'chlorides','free sulfur dioxide','total sulfur dioxide',
      'density','pH','sulphates','alcohol']]
y2=df[['quality']]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,test_size=.30,random_state=0)
```

## Creating Classifier Model

```python
classifier2=LogisticRegression(max_iter=1000)
classifier2.fit(x2_train.values,y2_train)
```

```python
y2_pred=classifier2.predict(x2_test.values)
```

```python
y2_pred[:10]
```

```python
y2_test.head(10)
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
cm2 = confusion_matrix(y2_test, y2_pred)
print(cm2)
```

```python
display2=ConfusionMatrixDisplay(confusion_matrix=cm2, display_labels=classifier2.classes_)
display2.plot()
plt.show()
```
